```python
import random
import math
import datetime
import pandas as pd
import matplotlib.pyplot as plt

# ==========================================
# PHẦN I: BÀI TẬP CƠ BẢN (Bài 1 - Bài 14)
# ==========================================

print("--- Bài 1: Toán tử cơ bản ---")
a, b = 10, 5 
print(f"Tổng: {a + b} | Hiệu: {a - b} | Tích: {a * b}")
if b != 0:
    print(f"Thương: {a / b}")
else:
    print("Không thể chia cho 0!")

print("\n--- Bài 2: Cắt chuỗi ---")
chuoi_1 = "Hello World"
ket_qua_cat = chuoi_1[2:5] 
print(f"Các ký tự từ chỉ số 2 đến 4: {ket_qua_cat}")

print("\n--- Bài 3: Xử lý chuỗi ---")
chuoi_2 = " Hello World "
chuoi_chuan = chuoi_2.strip()
print(f"Xóa khoảng trắng: '{chuoi_chuan}'")
print(f"In hoa: {chuoi_chuan.upper()}")
print(f"Thay 'H' thành 'J': {chuoi_chuan.replace('H', 'J')}")

print("\n--- Bài 4, 5, 6, 7, 8: Cấu trúc điều kiện ---")
if a > b: print("Bài 4: Hello World!")
if a == b:
    print("Bài 5: Yes")
else:
    print("Bài 5: No")

# Giả sử cho bài 6, 7, 8
c, d = 3, 5
if a == b: print("Bài 6: 1")
elif a > b: print("Bài 6: 2")
else: print("Bài 6: 3")

if a == b and b == d: print("Bài 7: Hello World!")
if a == b or c == d: print("Bài 8: Hello World!")

print("\n--- Bài 9 & 10: If...else rút gọn ---")
max_val = a if a > b else b
print(f"Max giữa {a} và {b} là: {max_val}")

# Bài 10: Input người dùng
# a_in = int(input("Nhập số a: "))
# b_in = int(input("Nhập số b: "))
# print("A") if a_in > b_in else print("=") if a_in == b_in else print("B")

print("\n--- Bài 11: List Comprehension ---")
n_ele = 5 # Ví dụ 5 phần tử
a_arr = [random.randint(1, 100) for _ in range(n_ele)]
b_arr = [x for x in a_arr if x % 2 == 0]
print(f"Mảng a: {a_arr}\nMảng b (chẵn): {b_arr}")

print("\n--- Bài 12: Bội số của 3 hoặc 5 ---")
tong_boi = sum(x for x in range(1000) if x % 3 == 0 or x % 5 == 0)
print(f"Tổng các số < 1000 là bội của 3 hoặc 5: {tong_boi}")

print("\n--- Bài 13: Trộn mảng tính tổng ---")
def tron_mang(mang1, mang2):
    ket_qua = []
    max_len = max(len(mang1), len(mang2))
    for i in range(max_len):
        val1 = mang1[i] if i < len(mang1) else 0
        val2 = mang2[i] if i < len(mang2) else 0
        ket_qua.append(val1 + val2)
    return ket_qua

print(f"Kết quả trộn mảng (sum): {tron_mang([3, 9, 1, 4], [2, 7, 4, 3, 2, 8])}")

print("\n--- Bài 14: Thao tác Ma trận ---")
m, n = 3, 3 # Ví dụ ma trận 3x3
matrix = [[random.randint(1, 10) for _ in range(n)] for _ in range(m)]
for row in matrix: print(row)

# d. Dòng có tổng lớn nhất
max_sum_row = max(matrix, key=sum)
print(f"Dòng có tổng lớn nhất: {max_sum_row}")

# h. TBC các phần tử biên
pt_bien = []
for i in range(m):
    for j in range(n):
        if i == 0 or i == m - 1 or j == 0 or j == n - 1:
            pt_bien.append(matrix[i][j])
print(f"TBC các phần tử biên: {sum(pt_bien)/len(pt_bien) if pt_bien else 0}")

# ==========================================
# PHẦN II: LỚP SINH VIÊN (Bài 15)
# ==========================================

print("\n--- Bài 15: Quản lý Sinh viên ---")
class SinhVien:
    def __init__(self, ma_sv, ten, nam_sinh, diem_tb):
        self.ma_sv = str(ma_sv)[:10]
        self.ten = str(ten)[:20]
        self.nam_sinh = int(nam_sinh)
        self.diem_tb = float(diem_tb)

danh_sach_sv = [
    SinhVien("02DH0001", "Phan Thi Lan", 2004, 7.5),
    SinhVien("01CD0002", "Nguyen Van A", 2007, 4.0),
    SinhVien("03DH0003", "Phan Van B", 2003, 8.0),
    SinhVien("04DH0004", "Le Thi Lan", 2006, 5.5)
]

def dem_sv_len_lop(ds):
    return sum(1 for sv in ds if sv.diem_tb >= 5)

def dem_sv_ho_phan(ds):
    # Sửa lỗi thụt lề tại đây
    return sum(1 for sv in ds if sv.ten.split()[0].upper() == "PHAN")

def dem_sv_ten_lan(ds):
    return sum(1 for sv in ds if sv.ten.split()[-1].upper() == "LAN")

print(f"Số SV đủ điều kiện lên lớp: {dem_sv_len_lop(danh_sach_sv)}")
print(f"Số SV họ Phan: {dem_sv_ho_phan(danh_sach_sv)}")
print(f"Số SV tên Lan: {dem_sv_ten_lan(danh_sach_sv)}")

# ==========================================
# PHẦN III: BÀI TẬP VỀ NHÀ
# ==========================================

print("\n--- BT VN 1: Trộn mảng lấy Min ---")
def tron_mang_min(a_m, b_m):
    ket_qua = []
    min_len = min(len(a_m), len(b_m))
    for i in range(min_len):
        ket_qua.append(min(a_m[i], b_m[i]))
    
    if len(a_m) > len(b_m):
        ket_qua.extend(a_m[min_len:])
    else:
        ket_qua.extend(b_m[min_len:])
    return ket_qua

print(f"Kết quả trộn mảng (min): {tron_mang_min([3, 9, 1, 4], [2, 7, 4, 3, 2, 8])}")

print("\n--- BT VN 2: Kiểm tra Ma trận ---")
def cheo_phu_giam_dan(a_mtx, n_sz):
    for i in range(n_sz - 1):
        if a_mtx[i][n_sz - 1 - i] <= a_mtx[i + 1][n_sz - 2 - i]:
            return False
    return True # Sửa lỗi thụt lề tại đây

def kiem_tra_doi_xung(a_mtx, n_sz):
    for i in range(n_sz):
        for j in range(i, n_sz):
            if a_mtx[i][j] != a_mtx[j][i]:
                return False
    return True

print("\n--- BT VN 4: File I/O ---")
with open("output_test.txt", "w", encoding="utf-8") as f:
    f.write("Dữ liệu thử nghiệm hệ thống.\nKết thúc bài tập.")
print("Đã ghi file 'output_test.txt' thành công.")

print("\n--- BT VN 5: Pandas & Matplotlib ---")
# Lưu ý: Sẽ hiện cửa sổ biểu đồ khi chạy
data_points = {
    'Môn học': ['CSDL', 'HĐH', 'AI', 'Xác Suất'],
    'Điểm Số': [8.5, 7.0, 9.0, 6.5]
}
df = pd.DataFrame(data_points)
print(df)

plt.bar(df['Môn học'], df['Điểm Số'], color='skyblue')
plt.title('Biểu đồ điểm số')
plt.ylim(0, 10)
# plt.show() # Bỏ comment nếu muốn xem biểu đồ ngay
```
